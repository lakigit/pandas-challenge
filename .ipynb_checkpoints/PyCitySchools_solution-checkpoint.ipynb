```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

```python
school_data_complete.count()
```

```python
#total schools
Total_schools = school_data_complete["school_name"].unique()
#len(Total_schools)
```

```python
#total number of students
Total_students = school_data_complete["student_name"].count()
#Total_students
```

```python
#total budget
Total_budget = school_data_complete["budget"].unique()
#Total_budget.sum()
```

```python
#average mathscore
average_math_score = school_data_complete["maths_score"].mean()
#average_math_score
```

```python
#average readingscore
average_reading_score = school_data_complete["reading_score"].mean()
#average_reading_score
```

```python
#Calculate the percentage of students with a passing maths score (50 or greater)
maths_pass = school_data_complete.loc[school_data_complete["maths_score"]>=50]
#maths_pass["maths_score"].count()
maths_pass_percentage = (maths_pass["maths_score"].count()/Total_students)*100
#maths_pass_percentage
```

```python
#Calculate the percentage of students with a passing reading score (50 or greater)
reading_pass = school_data_complete.loc[school_data_complete["reading_score"]>=50]
#maths_pass["reading_score"].count()
reading_pass_percentage = (reading_pass["reading_score"].count()/Total_students)*100
#reading_pass_percentage
```

```python
#Calculate the percentage of students who passed maths and reading (% Overall Passing)
maths_and_reading_pass = school_data_complete.loc[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]
#maths_and_reading_pass["Student ID"].count()
overall_pass = (maths_and_reading_pass["Student ID"].count()/Total_students)*100
#overall_pass
```

```python
#Create a dataframe to hold the above results
area_summary_df = pd.DataFrame({"Total Schools":[len(Total_schools)],"Total Students":[Total_students],"Total Budget":[Total_budget.sum()],
                           "Average Maths Score":[average_math_score],"Average Reading Score":[average_reading_score],"% Passing Maths":[maths_pass_percentage],
                          "% Passing Reading":[reading_pass_percentage],"% Overall Passing":[overall_pass]})
area_summary_df
```

```python
area_summary_df["Total Students"] = area_summary_df["Total Students"].map("{:,}".format)
area_summary_df["Total Budget"] = area_summary_df["Total Budget"].map("${:,.2f}".format)
area_summary_df
```

```python
#grouped = school_data_complete.groupby(["school_name"])
#grouped.mean()
```

```python
# School Type
each_school_types = school_data.set_index(["school_name"])["type"]
#each_school_types
```

```python
#Total Students
perschool_total_students = school_data_complete["school_name"].value_counts()
#perschool_total_students
```

```python
#Total School Budget
per_school_budget = school_data_complete.groupby(["school_name"])["budget"].mean()
#per_school_budget
```

```python
#Per Student Budget
per_student_budget = per_school_budget/perschool_total_students
#per_student_budget
```

```python
#Average Maths Score
per_maths_average = school_data_complete.groupby(["school_name"])["maths_score"].mean()
#per_maths_average
```

```python
#Average Reading Score
per_reading_average = school_data_complete.groupby(["school_name"])["reading_score"].mean()
#per_reading_average
```

```python
#% Passing Maths
total_maths_pass = school_data_complete.loc[school_data_complete["maths_score"]>=50]
maths_pass_per_school = total_maths_pass.groupby(["school_name"]).count()["Student ID"]
#maths_pass_per_school
maths_pass_percentage = (maths_pass_per_school/perschool_total_students)*100
#maths_pass_percentage
```

```python
#% Passing Reading
total_reading_pass = school_data_complete.loc[school_data_complete["reading_score"]>=50]
reading_pass_per_school = total_reading_pass.groupby(["school_name"]).count()["Student ID"]
#reading_pass_per_school
reading_pass_percentage = (reading_pass_per_school/perschool_total_students)*100
#reading_pass_percentage
```

```python
#% Overall Passing (The percentage of students that passed maths and reading.)
total_maths_and_reading_pass = school_data_complete.loc[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]
pass_per_school = total_maths_and_reading_pass.groupby(["school_name"]).count()["Student ID"]
perschool_pass_percentage = (pass_per_school/perschool_total_students)*100
#perschool_pass_percentage
```

```python
school_summary = pd.DataFrame({"School Type":each_school_types,
                               "Total Students":perschool_total_students,
                               "Total School Budget":per_school_budget,
                               "Per Student Budget":per_student_budget,
                              "Average Maths Score":per_maths_average,
                              "Average Reading Score":per_reading_average,
                              "% Passing Maths":maths_pass_percentage,
                              "% Passing Reading":reading_pass_percentage,
                              "% Overall Passing":perschool_pass_percentage})
school_summary
```

```python
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].str.replace("$","",regex=False).astype("float")
school_summary.head()
```

```python
spending_bins = [0, 585, 630, 645, 680]
group_name = ["<$585", "$585-630", "$630-645", "$645-680"]
school_summary["Spending Ranges (Per Student)"] =pd.cut(school_summary["Per Student Budget"],spending_bins,labels = group_name, include_lowest=True)
school_summary
```

```python
spending_summary_group = school_summary.groupby(["Spending Ranges (Per Student)"])
spending_math_scores = spending_summary_group["Average Maths Score"].mean()
spending_reading_scores = spending_summary_group["Average Reading Score"].mean()
spending_passing_math = spending_summary_group["% Passing Maths"].mean()
spending_passing_reading = spending_summary_group["% Passing Reading"].mean()
overall_passing_spending = spending_summary_group["% Overall Passing"].mean()
```

```python
spending_summary = pd.DataFrame({"Average Maths Score":spending_math_scores,"Average Reading Score":spending_reading_scores,
                                 "% Passing Maths":spending_passing_math,"% Passing Reading":spending_passing_reading,
                                "% Overall Passing":overall_passing_spending})
spending_summary
```

```python
spending_summary["Average Maths Score"] = spending_summary["Average Maths Score"].map("{:.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:.2f}".format)
spending_summary["% Passing Maths"] = spending_summary["% Passing Maths"].map("{:.2f}".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].map("{:.2f}".format)
spending_summary["% Overall Passing"] = spending_summary["% Overall Passing"].map("{:.2f}".format)
spending_summary
```

```python
size_summary = pd.DataFrame({"Average Maths Score":size_math_scores,"Average Reading Score":size_reading_scores,
                                 "% Passing Maths":size_passing_math,"% Passing Reading":size_passing_reading,
                                "% Overall Passing":overall_passing_size})
size_summary
```

```python
size_summary["Average Maths Score"] = size_summary["Average Maths Score"].map("{:.2f}".format)
size_summary["Average Reading Score"] = size_summary["Average Reading Score"].map("{:.2f}".format)
size_summary["% Passing Maths"] = size_summary["% Passing Maths"].map("{:.2f}".format)
size_summary["% Passing Reading"] = size_summary["% Passing Reading"].map("{:.2f}".format)
size_summary["% Overall Passing"] = size_summary["% Overall Passing"].map("{:.2f}".format)
size_summary
```
